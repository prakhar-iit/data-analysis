```python
def find_Question_Data(testname, section_Name):
    body = driver.page_source
    soup = BeautifulSoup(body)
    table = soup.find_all("div", {"class": "search-table-outter"})
    rows = table[0].find_all("tr")
    parsed_table_data = []
    for row in rows: 
        children = row.findChildren(recursive=False)
        row_text = []
        for child in children:
            clean_text = child.text
            clean_text = clean_text.split('&#91;')[0]
            clean_text = clean_text.split('&#160;')[-1]
            clean_text = clean_text.strip()
            row_text.append(clean_text)
        parsed_table_data.append(row_text)
    question_Data = [testname, section_Name, '', parsed_table_data[1][2], parsed_table_data[1][3], '', parsed_table_data[2][2], parsed_table_data[2][3], parsed_table_data[3][1], parsed_table_data[3][2], parsed_table_data[3][3], parsed_table_data[4][2], parsed_table_data[4][3]]
    return question_Data
    
```

```python
def find_VARC_Data(df, testName, section_Name, vaQuestions, driver):
    for x in range(vaQuestions):
        question_data = find_Question_Data(testName, section_Name); 
        df.loc[len(df.index)] = question_data
        #Go to next Questions
        if(x!=vaQuestions-1):
            next_button = driver.find_element_by_id("next"); 
            next_button.click(); 
```

```python
def find_DILR_Data(df, testName, section_Name, dilrQuestions, driver):
    for x in range(dilrQuestions):
        question_data = find_Question_Data(testName, section_Name); 
        df.loc[len(df.index)] = question_data
        #Go to next Questions
        if(x!=dilrQuestions-1):
            next_button = driver.find_element_by_id("next"); 
            next_button.click();     
```

```python
def find_QA_Data(df, testName, section_Name, qaQuestions , driver):
    for x in range(qaQuestions):
        question_data = find_Question_Data(testName, section_Name); 
        df.loc[len(df.index)] = question_data
        #Go to next Questions
        if(x!=qaQuestions-1):
            next_button = driver.find_element_by_id("next"); 
            next_button.click(); 
```

```python
def find_Test_Data(testName, driver, vaQuestions, dilrQuestions, qaQuestions): 
    df = pd.DataFrame(columns=['Test Name', 'Section Name', 'My Attempt (Y/N)','Overall Attempt','Toppers Attempt','Status (R/W)','Overall Accuracy','Toppers Accuracy','Time Taken', 'Overall Time Taken', 'Toppers Time Taken', 'Overall P Value', 'Toppers P Value'])
    VARC_data = find_VARC_Data(df, testName, "VARC", vaQuestions, driver); 
    change_Section(1);
    DILR_data = find_DILR_Data(df, testName, "DILR", dilrQuestions, driver); 
    change_Section(2);
    QA_data = find_QA_Data(df, testName, "QA",qaQuestions, driver); 
    return df;
```

```python
def change_Section(index):
    change_Section = driver.find_element_by_id("sectionblock")
    soup = BeautifulSoup(change_Section.get_attribute('innerHTML'))
    available_sections = soup.find_all("div")
    time.sleep(2); 
    driver.find_element_by_xpath('//*[@id="testScreen"]/div/div[1]/div').click() 
    driver.find_element_by_id(available_sections[index].get('id')).click()
```
